```python
import pandas as pd

#3.2.1
df = pd.read_csv("titanic.csv")  #loads csv file
df.info()  #checks dataframe info

df["Age"] = df["Age"].fillna(df["Age"].median())   #fillna() replaces all the empty cells with a given value. i.e median value in this case
df["fare_per_year"] = df["Fare"] / df["Age"] #new colums which contains fare divided by age of passenger

high_fare_age = df[df["fare_per_year"] > 5]  #fare_per_year > 5 bhako ma rows subset gareko
high_fare_age_srt = high_fare_age.sort_values(by="fare_per_year", ascending=False)  #high_fare_age and descending sorting
result = high_fare_age_srt[["Name", "fare_per_year"]] #high_fare_age_srt ma name ra fare_per_year matra rakhera result ma store gareko
print(result)  # Allison, Master. Hudson Trevor had the highest fare relative to age, with a fare per year of approximately 164.73


df["fare_per_class"] = df["Fare"] / df["Pclass"] # creating a new column calculating fare per class
adult_males = df[(df["Sex"] == "male") & (df["Age"] >= 18)] # calculating adult males
adult_males_srt = adult_males.sort_values(by="fare_per_class", ascending=False) # sorting adult males by the valujes of fare per class and making the order descending
result = adult_males_srt[["Name", "Age", "fare_per_class"]] # adult_males_srt ma bhayeko name age ani fare perclass lai result ma store gareko
print(result) # Lesurer, Mr. Gustave J and Cardeza, Mr. Thomas Drake Martinez paid the highest fare relative to their class, with a fare per class of approximately 512.33


```

```python
import pandas as pd

#3.3
df = pd.read_csv("titanic.csv")
df.info()
total_fare = df['Fare'].sum()  # total paid fare by all passengers ko sum
first_class = df[df["Pclass"] == 1] # passenger class 1 bhayeko lai filter garera chuttako
total_fare_first_class = first_class["Fare"].sum() # pclass one bhayeko haru le total paid fare ko sum
second_class = df[df["Pclass"] == 2] # same as pclass 1
total_fare_second_class = second_class["Fare"].sum() # same as pclass 1
third_class = df[df["Pclass"] == 3]   # same as pclass 1
total_fare_third_class = third_class["Fare"].sum()   # same as pclass 1
fare_totals_by_class = [   # sabai pclass ko total paid fare lai list create garera rakheko chuttai chuttai
    total_fare_first_class,
    total_fare_second_class,
    total_fare_third_class
]
fare_proportions = [   # first,second ra third class ko total paid fare haruko proportion nikaleko
    total_fare_first_class / total_fare,
    total_fare_second_class / total_fare,
    total_fare_third_class / total_fare
]
print(fare_proportions)
#63.3% of total fare was contributed by first class passangers
#13.2% of the total fare was contibuted by second class passangers
#23.4% of the total fare was contributed by third class passangers
df["Age"] = df["Age"].fillna(df["Age"].median())  # null wa zero value bhayeko thaun ma age ko median le replace gareko

df.loc[df["Age"] < 18, "age_group"] = "child" # age 18 bhanda less bhayeko haru lai child bhanne age group ma rakheko
df.loc[(df["Age"] >= 18) & (df["Age"] < 65), "age_group"] = "adult" # 18 bhanda older but 65 bhanda younger lai adult ma rakheko
df.loc[df["Age"] >= 65, "age_group"] = "senior" # baki lai senior ma
total_passengers = len(df)  # total passengers ko count
age_group_counts = df["age_group"].value_counts() # calculation of total passenger ko numbers for creating proportion
age_group_proportions = age_group_counts / total_passengers # adults, childrens ra seniors ko proportion on filling the train i guess
age_group_percent = age_group_proportions * 100
print(age_group_percent) #approx 86.10% were adults, 12.69% of passangers were children and 1.23% were seniors
```
